Longest Proper Prefix which is also Suffix

Each lps[i] tells us:
The length of the longest proper prefix which is also a suffix for the substring pattern[0..i].

▶️ Step-by-Step Explanation for "ababaca"
Let’s go index by index:

Index 0 → "a"
Substring: "a"

No proper prefix or suffix

lps[0] = 0

Index 1 → "ab"
Prefix: "a"

Suffix: "b"

No match

lps[1] = 0

Index 2 → "aba"
Prefixes: "a", "ab"

Suffixes: "a", "ba"

Common: "a"

lps[2] = 1

Index 3 → "abab"
Prefixes: "a", "ab", "aba"

Suffixes: "b", "ab", "bab"

Common: "ab"

lps[3] = 2

Index 4 → "ababa"
Prefixes: "a", "ab", "aba", "abab"

Suffixes: "a", "ba", "aba", "baba"

Common: "aba"

lps[4] = 3

Index 5 → "ababac"
Prefixes: "a", "ab", "aba", "abab", "ababa"

Suffixes: "c", "ac", "bac", "abac", "babac"

No match

lps[5] = 0

Index 6 → "ababaca"
We just had a mismatch at index 5, so we restart the check.

Check if pattern[6] == pattern[0] → "a" == "a" → match!

So lps[6] = 1



```python
pattern = 'abababcb'
length = 0
n =len(pattern)
i = 1
lps = [0] * n
while i < n:
    if pattern[i] == pattern[length]:
        length +=1
        lps[i] = length
        i +=1
    else:
        if length != 0:
            length = lps[length - 1]
        else:
            lps[i] = 0
            i +=1
print(f"lps of pattern {pattern} is {lps}")
```
